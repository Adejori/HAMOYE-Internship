```python
import pandas as pd
```

```python
energydata = pd.read_csv('C:/Users/HP/Desktop/Hamoye Internship/STAGE B/Quiz dataset/energydata_complete.csv')
energydata
```

```python
energy_data = energydata.drop(columns=['date','lights'])
energy_data
```

```python
energy_data.info()
```

```python
feature_col = energy_data.drop(columns = ["T2", "T6"])
target_col = energy_data["T6"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_col,target_col, test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

QUESTION 12 AND 16

```python
predicted_values = linear_model.predict(X_test)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

QUESTION 13

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae,2)
```

QUESTION 14

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss,2)
```

QUESTION 15

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

QUESTION 17

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_model_weights = get_weights_df(linear_model, X_train, 'linear_Model_Weight')
linear_model_weights
```
